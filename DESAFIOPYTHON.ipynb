```python
# a quantidade de funcionários por cada setor
df2 = df.groupby('Department')[['EmployeeNumber']].count()
df2
```

```python
df2.plot(kind= 'bar')
```

```python
# Quantidade de funcionário por cada nível de escolaridade
df3 = df.groupby('Education')[['EmployeeNumber']].count()
df3
```

```python
df3.plot(kind = 'line', marker = 'o')
```

```python
# Quantidade de funcionários com a coluna de atrito na empresa
df4 = df.groupby('Attrition')[['EmployeeNumber']].count()
df4
```

```python
df4.plot(kind = 'pie', subplots = True)
```
